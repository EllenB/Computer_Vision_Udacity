## Import the relevant libraries

You need to first select four points and then select where the four points are.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#%matplotlib inline

# Read in the image
image = cv2.imread('C:\Ellen\Udacity\Computer_Vision\data\example_01.png')
image.shape
```

```python
# Make a copy of the card
image_copy = np.copy(image)

# Change color to RGB (from GBR)
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
```

```python
#plt.imshow(image_copy)
```

Create 4 points from the source image but use the coordinates going counter clockwise starting at the top left. Typing in the following command opens up the image in an interactive way such that you can see the corner coordinates.

```python
%matplotlib

plt.imshow(image_copy)
```

```python
%matplotlib inline
plt.imshow(image_copy)

# Type in the coordinates of the four points
plt.plot(72,234,'.') # Top left
plt.plot(184,436,'.') # bottom left
plt.plot(469,257,'.') # bottom right
plt.plot(352,110,'.') # top right
```

```python
## Define a function called warp

def warp(image):
    
    # Four source coordinates, which define a rectangular plane
    source_pts = np.float32(
    [[72,234],
    [184,436],
    [469,257],
    [352,110]])
    
    # Four warped coordinates
    warped_pts = np.float32(
    [[100,200],
    [100,400],
    [400,400],
    [400,200]])
    
    # Compute the perspective transform, M:
    M = cv2.getPerspectiveTransform(source_pts,warped_pts)
    
    # Can compute the inverse by swapping the input parameters
    M_inv = cv2.getPerspectiveTransform(warped_pts,source_pts)
    
    # Get the image size
    image_size = (image.shape[1],image.shape[0])
    
    # Computes and returns the warped image
    warped = cv2.warpPerspective(image,M,image_size, flags = cv2.INTER_LINEAR)
    
    return warped
```

## Compute and plot the transformed image

```python
%matplotlib inline
warped_image = warp(image_copy)

# Create side-by-side plots

f, (ax1, ax2) = plt.subplots(1,2,figsize = (20,10))

ax1.set_title('Source image')
ax1.imshow(image_copy)

ax2.set_title('Warped image')
ax2.imshow(warped_image)
```
