## Import resources and display image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('C:\\Ellen\\Udacity\\Computer_Vision\\data\\city-hall.png')

# Make a copy of the image
image_copy = np.copy(image)

# Change color of the image to RGB
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

## Convert to grayscale

```python
gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)

plt.imshow(gray,cmap='gray')
```

## Create a customer kernel

In order to detect vertical edges rather than horizontal edges. Compute differences between left and right.

This filter is actually called the Sobel filter. 

```python
# 3 x 3 array for edge detection

sobel_x = np.array([[-1,0,1],
                 [-2,0,2],
                 [-1,0,1]])
```

Then, use this filter using a function in Open CV:

```python
filtered_image = cv2.filter2D(gray, -1, sobel_x) #-1 means the same output type
plt.imshow(filtered_image)
```

```python
## Create into a binary image
retval,binary_image = cv2.threshold(filtered_image, 100,255,cv2.THRESH_BINARY)

plt.imshow(binary_image,cmap = 'gray')
```

There are dots on the grass that do not provide any extra information. With a high-pass filter, noise is introduced. With a low-pass filter can reduce noise and you need to do before doing the high pass edge detection filter.
