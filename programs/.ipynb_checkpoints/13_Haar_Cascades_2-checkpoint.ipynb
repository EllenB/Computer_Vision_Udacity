## Import the relevant libraries

```python
import cv2

import numpy as np

import matplotlib.pyplot as plt

from common import clock, mosaic

%matplotlib inline
```

## One face

You can find the Haar cascades here: C:\opencv\sources\data\haarcascades

```python
#setting up the haar cascade classifiers from the opencv installation
face_cascade = cv2.CascadeClassifier('C:\\Ellen\\Udacity\\Computer_Vision\\data\\haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('C:\\Ellen\\Udacity\\Computer_Vision\\data\haarcascade_eye.xml')

face_cascade
eye_cascade
```

## Face and eye detection

```python
img = cv2.imread('C:\\Ellen\\personal\\PassportPhotos\\Ellen-Photo.jpg')
# Print out the shape of the image
print('Shape image:')
print(img.shape)
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap ='gray')
```

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

```python
#search for faces
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
```

```python
#for each face, detect eyes
print(len(faces))
for (x,y,w,h) in faces:
    cv2.rectangle(img_RGB,(x,y),(x+w,y+h),(255,0,0),2) # Make this a red box and 2 stands for the line width
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img_RGB[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2) # Make this a green box
        
plt.imshow(img_RGB)
plt.show()
plt.close()
```

## Multiple faces and eyes

```python
img = cv2.imread('C:\\Ellen\\Udacity\\Computer_Vision\\data\\Family.jpg')
# Print out the shape of the image
print('Shape image:')
print(img.shape)
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap ='gray')
```

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

```python
#search for faces
faces = face_cascade.detectMultiScale(gray, 1.3, 2)
```

```python
#for each face, detect eyes
print(len(faces))
for (x,y,w,h) in faces:
    cv2.rectangle(img_RGB,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img_RGB[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        
plt.imshow(img_RGB)
plt.show()
plt.close()
```
