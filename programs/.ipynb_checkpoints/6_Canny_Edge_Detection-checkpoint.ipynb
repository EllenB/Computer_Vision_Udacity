## Import resources and display image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

Read in the image:

```python
image = cv2.imread('C:\\Ellen\\Udacity\\Computer_Vision\\data\\sunflower.jpg')
```

```python
# Make a copy:
```

```python
image_copy = np.copy(image)
```

Change color to RGB:

```python
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

## Implement canny edge detection

Define lower and upper thresholds for hysteresis

```python
lower = 120
upper = 240

edges=cv2.Canny(gray,lower,upper)

edges = plt.imshow(edges,cmap ='gray')
```

Try different settings:

```python
wide = cv2.Canny(gray,30,100)
tight = cv2.Canny(gray,180,240)
#plt.imshow(wide,cmap='gray')

# Display the images
f,(ax1, ax2) = plt.subplots(1,2,figsize=(20,10))

ax1.set_title('wide')
ax1.imshow(wide,cmap='gray')
              
ax2.set_title('tight')
ax2.imshow(tight,cmap='gray')              
              
```
