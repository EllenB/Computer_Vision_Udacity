```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
%matplotlib inline
```

```python
# Read the image
image = cv2.imread('C:\\Ellen\\Udacity\\Computer_Vision\\data\\phone.jpg')
image.shape
```

```python
# Make a copy of the image
image_copy = np.copy(image)

# Change color to RGB
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

Image we want so isolate the screen area. First we need to detect the edges. First, we convert to grayscale and then we apply a threshold and apply the Canny edge detector.

```python
# Convert image to grayscale
gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)

# Define our parameters for Canny
low_threshold = 50
high_threshold = 100

edges =  cv2.Canny(gray, low_threshold,high_threshold)

plt.imshow(edges,cmap='gray')
```

```python
# Find lines through a Hough transform

# Parameters to take into the function
rho = 1 # Resolution of the degree is one pixel as rho is measured in pixels
theta = np.pi/180 # one degree
threshold = 50 # Minimum vote it should get to be considered a line (number of votes)
min_line_length = 40
max_line_gap = 7

lines =cv2.HoughLinesP(edges,rho,theta,threshold,np.array([]), min_line_length,max_line_gap)

line_image = np.copy(image_copy)

# Iterate over the output "lines" and draw lines on the image copy
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),5)
        
plt.imshow(line_image)
```

What has been taken is a probabalistic Hough Transform. The reason for doing so is that considering all the points can be very computer intensive. Instead, only a random subset of points as this should be sufficient for line detection.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Read in an image and convert to RGB
image = cv2.imread('C:\\Ellen\\Udacity\\Computer_Vision\\data\\road_1.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Peform edge detection with Canny
low_threshold = 90
high_threshold = 180
edges = cv2.Canny(gray, low_threshold, high_threshold)
# ---------------------------------------------------------- #


## TODO: Define the Hough transform parameters 
## so that lane lines are the only ones detected
rho = 1
theta = np.pi/180
threshold = 50
min_line_length = 60
max_line_gap = 5


line_image = np.copy(image) #creating an image copy to draw lines on

# Run Hough on the edge-detected image
lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]),
                        min_line_length, max_line_gap)


# Iterate over the output "lines" and draw lines on the image copy
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),3)

# ---------------------------------------------------------- #
plt.imshow(line_image)

```
